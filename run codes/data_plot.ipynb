```python
import seaborn as sns
import matplotlib.pyplot as plt

model = [0.0, 12.0, 16.0, 17.0, 28.0, 32.0, 26.0, 39.0, 48.0, 61.0, 64.0, 61.0, 41.0, 41.0, 56.0, 80.0, 76.0]
real = [0.0, 9.0, 14.0, 34.0, 34.0, 44.0, 50.0, 43.0, 41.0, 44.0, 76.0, 33.0, 69.0, 63.0, 76.0, 78.0, 77.0]

# Create a DataFrame for the data
import pandas as pd
df = pd.DataFrame({'Index': range(1, len(model) + 1), 'Model': model, 'Real': real})

# Set the style
sns.set(style="whitegrid")

# Create a line plot using seaborn with x-axis increasing by 1
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Index', y='Model', label='Model', marker='o')
sns.lineplot(data=df, x='Index', y='Real', label='Real', marker='o')

# Set labels and title
plt.xlabel('Rounds')
plt.ylabel('Objective Value')
plt.title('Comparison of Model and Real Schedules (Bundesliga)')

# Show legend
plt.legend()

# Show the plot
plt.show()
```
