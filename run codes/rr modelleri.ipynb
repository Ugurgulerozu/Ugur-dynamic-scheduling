```python
from pyomo.environ import *
import pandas as pd
import math
import numpy as np

#Tankut hocanın yazdığı modeli kodladım

n = 4

teams = list(range(1,n+1))
days = list(range(1,n))

# Create ConcreteModel
model = ConcreteModel()
# Variables
model.x = Var(teams, teams, days, within=Binary)



# Constraints
model.constraints = ConstraintList()



for i in teams:
    for j in teams:
        if i<j:
            model.constraints.add(sum(model.x[i,j,t] for t in days) == 1)
        else :
            model.constraints.add(sum(model.x[i,j,t] for t in days) == 0)


for i in teams:
    for t in days:
            model.constraints.add(sum(model.x[i,j,t] + model.x[j,i,t] for j in teams if i!=j) == 1)

def rule_of(model):
    return 0

model.obj = Objective(rule=rule_of, sense=minimize)



# Solve the model
solver = SolverFactory('gurobi', options={'MIPFocus':2, 'Heuristics':1}) 
results= solver.solve(model, tee=True)
obj_value = value(model.obj())
#model.write(f"my_model6.lp", io_options={'symbolic_solver_labels': True})
model.constraints.pprint()


if results.solver.termination_condition == TerminationCondition.optimal:
    print("Optimal solution found.")
    for t in days:
        for i in teams:
            for j in teams:
                    if value(model.x[i, j, t]) == 1:
                        print(i,j,t)
                        print(f"Team {i} vs Team {j} in Slot {t}")


else:
    print("No optimal solution found.")
```
